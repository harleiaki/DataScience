```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('SINASC_RO_2019.csv', sep=',', decimal=',')
df.head()
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
df[['IDADEMAE']].boxplot()
sns.boxplot(y="IDADEMAE", data=df[df['IDADEMAE']>1])
```

####  2) Faça um boxplot do peso do bebê

```python
df[['PESO']].boxplot()
sns.boxplot(y='PESO', data=df[df['PESO']>1])
```

####  3) Faça um histograma da idade da mãe

```python
df['IDADEMAE'].plot.hist(bins = 20)
```

```python
sns.displot(data = df[df['IDADEMAE']>1]
            , alpha = .50  #cor das barras
            , x = 'IDADEMAE' # variavel 
            , bins = 20 # numero de quebras igual do histograma
            , element = 'step' # linhas verticais das barras
            , kde = True # linha da estimativa 
            )
```

#### 4) Faça um histograma do peso do bebê

```python
df['PESO'].plot.hist(bins=20)
```

```python
sns.displot(data = df[df['IDADEMAE']>0]
            , alpha = 1
            , x = 'PESO'
            , bins = 20
            #, element = 'step'
            #, kde = True
            )
```
