#### 3) Monte um gráfico de barras mostrando o APGAR5 médio por CONSULTAS

```python
plt.close('all')
sns.barplot(x='APGAR5', y='CONSULTAS', data=pd)
```

```python
plt.close('all')

plt.rc('figure', figsize=(15, 8))
fig, axes = plt.subplots(2, 3)

sns.countplot(ax = axes[0, 0], x='APGAR1', data=pd)
sns.countplot(ax = axes[0, 1], x='APGAR5', data=pd)
sns.countplot(ax = axes[0, 2], x='SEXO', data=pd)
gestacao = sns.countplot(ax = axes[1, 0], x='GESTACAO', data=pd)
sns.countplot(ax = axes[1, 1], x='ESTCIVMAE', data=pd)

gestacao.set_xticklabels(gestacao.get_xticklabels(), rotation=90, horizontalalignment='right')

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

####  4) Monte um painel com 4 gráficos mostrando as frequencias de SEXO, GESTACAO, ESTCIVMAE e CONSULTAS

```python
plt.close('all')

plt.rc('figure', figsize=(10, 10))
fig, axes = plt.subplots(2, 2)

sns.countplot(ax = axes[0, 0], x='SEXO', data=pd)
gestacao = sns.countplot(ax = axes[1, 0], x='GESTACAO', data=pd)
sns.countplot(ax = axes[0, 1], x='ESTCIVMAE', data=pd)
sns.countplot(ax = axes[1, 1], x='CONSULTAS', data=pd)

gestacao.set_xticklabels(gestacao.get_xticklabels(), rotation=90, horizontalalignment='right')

plt.subplots_adjust(wspace=0.2, hspace=0.2)
```
