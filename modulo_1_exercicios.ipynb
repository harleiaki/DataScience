## 4\. Booleanos

Utilize a tabela da verdade para responder: qual o valor da variável x?

```
a = False
b = True

x = not a & b
```

Aqui está a tabela da verdade para a expressão "x = not a & b":

| A        | B        | | | NOT A  | | NOT A & B  | | X    |
|----------|----------|-|-|--------|-|------------|-|----------|
| FALSE    | TRUE     | | | TRUE   | | TRUE       | | FALSE    |

Portanto, o valor da variável x é False.
